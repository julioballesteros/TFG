# 1. Análisis general de los datos

```python
# frecuenca del número de repeticiones (count) de los indicadores
display(variants_df['count'].value_counts())

# frecuencia relativa
display(variants_df['count'].value_counts() / len(variants_df['count']))

# crear gráfico del número de entradas por tipo
plt.figure(figsize=(20, 10))
sns.countplot(x='count', data=variants_df)
plt.title('Frecuencia de repetición de los indicadores')
plt.xlabel('Número de repeticiones')
plt.ylabel('Frecuencia')

plt.show()
```

# 2 Indicadores de tipo Hash

## 2.1 Análisis de los hashes

```python
# obtener dataset hashes
hashes_types = ['FileHash-SHA1', 'FileHash-SHA256', 'FileHash-MD5']
hashes_df = variants_df[ variants_df['indicator'].isin(hashes_types) ]
hashes_df = hashes_df.drop(columns='indicator')
hashes_df = hashes_df.rename(columns={'value': 'hash'})
```

```python
# examinar dataset
display(hashes_df.info())
display(hashes_df.head())
```

```python
# obtener número de hashes por variante
display(hashes_df['variant'].value_counts())

# crear gráfico del número de entradas por tipo
plt.figure(figsize=(20, 10))
sns.countplot(x='variant', data=hashes_df)
plt.title('Número de hashes por variante')
plt.xlabel('Variante')
plt.ylabel('Número de hashes')

plt.show()
```

```python
# obtener cuántos hashes se repiten en más de una variante
display(hashes_df['hash'].value_counts().value_counts())
```

```python
repeated_hashes = hashes_df['hash'].value_counts().index[hashes_df['hash'].value_counts() > 1]

for variant in hashes_df['variant'].unique():
    hashes_df[hashes_df['variant'] == variant] in repeated_hashes
```

```python
hashes_df[hashes_df['variant'] == 'okiru'] in repeated_hashes
```

# 3 Indicadores IPv4

## 3.1 Análisis IPs

```python
# obtener dataset de los indicadores de tipo IP
ip_df = variants_df[ variants_df['indicator'] == 'IPv4' ]
ip_df = ip_df.drop(columns='indicator')
ip_df = ip_df.rename(columns={'value':'IPv4'})
```

```python
# examinar dataset
display(ip_df.info())
display(ip_df.head())
```

```python
# obtener número de IPs por cada variante
display(ip_df['variant'].value_counts())

# crear gráfico del número de entradas por tipo
plt.figure(figsize=(20, 10))
sns.countplot(x='variant', data=ip_df)
plt.title('Número de IPs por variante')
plt.xlabel('Variante')
plt.ylabel('Número de IPs')

plt.show()
```

```python
# ver la frecuencia de repetición de IPs
display(ip_df['IPv4'].value_counts().value_counts())
```

## 3.2 Mapa de geolocalización de IPs

```python
# obtener localización de las direcciones IP
latitude = []
longitude = []
country_name = []
country_code = []

for ip in ip_df['IPv4']:
    response = requests.get("https://geolocation-db.com/json/" + ip + "&position=true").json()
    
    latitude.append(response['latitude'])
    longitude.append(response['longitude'])
    country_name.append(response['country_name'])
    country_code.append(response['country_code'])
    
ip_df['latitude'] = latitude
ip_df['longitude'] = longitude
ip_df['country_name'] = country_name
ip_df['country_code'] = country_code
```

```python
# convertir las coordenadas a números
lat = []
lng = []
bad_rows = []
for i in range(len(ip_df)):
    try:
        lat.append(float(ip_df.iloc[i]['latitude']))
        lng.append(float(ip_df.iloc[i]['longitude']))
    except:
        bad_rows.append(i)
        
# eliminar las filas con errores
ip_df = ip_df.drop(bad_rows)

ip_df['latitude'] = lat
ip_df['longitude'] = lng
```

```python
# examinar dataset
display(ip_df.info())
display(ip_df.head())
```

```python
# eliminar nans
ip_df = ip_df.dropna()
ip_df.info()
```

```python
# obtener dataset de mirai
mirai_df = ip_df.loc[ip_df.variant == 'mirai']

# obtener límites del mapa
llat = mirai_df['latitude'].min() + 2
ulat = mirai_df['latitude'].max() + 2
llon = mirai_df['longitude'].min() + 2
ulon = mirai_df['longitude'].max() + 2
```

```python
fig, ax = plt.subplots(figsize=(20, 10))

ip_map = Basemap(projection='merc', resolution='l', area_thresh=1000.0, llcrnrlon=llon, llcrnrlat=llat, urcrnrlon=ulon, urcrnrlat=ulat)
ip_map.drawcoastlines()
ip_map.drawcountries()
ip_map.drawlsmask(land_color='orange', ocean_color='skyblue')
xs, ys = ip_map(np.asarray(mirai_df['longitude']), np.asarray(mirai_df['latitude']))
mirai_df['x'] = xs.tolist()
mirai_df['y'] = ys.tolist()

for i, row in ip_df.iterrows():
    ip_map.plot(row.x, row.y, markerfacecolor='purple', markeredgecolor='white', marker='o', markersize=6)

plt.title('Location of Mirai IPs')

plt.savefig('IP_mirai_map.png', dpi=300)
plt.show()
```

```python
# obtener dataset de las variants
variants_df = ip_df.loc[ip_df.variant != 'mirai']

# obtener límites del mapa
llat = variants_df['latitude'].min() + 2
ulat = variants_df['latitude'].max() + 2
llon = variants_df['longitude'].min() + 2
ulon = variants_df['longitude'].max() + 2
```

```python
fig, ax = plt.subplots(figsize=(20, 10))

ip_map = Basemap(projection='merc', resolution='l', area_thresh=1000.0, llcrnrlon=llon, llcrnrlat=llat, urcrnrlon=ulon, urcrnrlat=ulat)
ip_map.drawcoastlines()
ip_map.drawcountries()
ip_map.drawlsmask(land_color='orange', ocean_color='skyblue')
xs, ys = ip_map(np.asarray(variants_df['longitude']), np.asarray(variants_df['latitude']))
variants_df['x'] = xs.tolist()
variants_df['y'] = ys.tolist()

cmap = ['purple', 'green', 'red', 'blue', 'black', 'white', 'yellow', 'grey']
variants = variants_df['variant'].unique()
colors = {}
for i, v in enumerate(variants):
    colors[v] = cmap[i]

for i, row in variants_df.iterrows():
    ip_map.plot(row.x, row.y, markerfacecolor=colors[row.variant], markeredgecolor='white', marker='o', markersize=10, alpha=1)

plt.title('Location of Mirai variants IPs')

plt.savefig('IP_variants_map.png', dpi=300)
plt.legend()
plt.show()
```
