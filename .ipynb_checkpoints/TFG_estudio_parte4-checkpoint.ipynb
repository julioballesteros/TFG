# 2. Predecir el precio de cierre
En esta sección se prueban modelos para predecir la serie temporal de los precios de cierre. Utilizando distintos conjuntos de variables para obtener el mejor resultado. Esto es la variable 'future_close', la cual es cuantitativa, continua y no estacionaria.

## 2.1. Serie temporal al dato siguiente

### Definir las variables necesarias para el modelo

```python
# variables independientes que se utilizan para la predicción
features = ['close']#, 'volume', 'adx', 'nvi', 'smad_10', 'smad_20', 'rsi_14', 'rsi_28', 'bbd', 'ppo']

# fecha de testeo
test_date = '2018-1-1'

# número de datos que se utilizan para predecir el siguiente
lookback = 15
```

### Preprocesado

```python
# obtener conjuntos de entrenamiento y testeo como array de numpy bidimensiaonal
# en testeo añadir ventana de predicción al principio
training_set = stock_prices.loc[:test_date, features].values
testing_set = np.concatenate((training_set[-lookback:],
                                  stock_prices.loc[test_date:, features].values))

# escalar variables
scaler = MinMaxScaler()
training_set = scaler.fit_transform(training_set)
testing_set = scaler.transform(testing_set)

# crear sliding window (ventana de predicción)
X_train = []
y_train = []
for i in range(lookback, len(training_set)):
    X_train.append(training_set[i-lookback: i, :])
    # en test se añade solo el precio de cierre (es el target)
    y_train.append(training_set[i, 0])
X_train, y_train = np.array(X_train), np.array(y_train)

# crear sliding window de testeo
X_test = []
y_test = []
for i in range(lookback, len(testing_set)):
    X_test.append(testing_set[i-lookback: i, :])
    y_test.append(testing_set[i, 0])
X_test, y_test = np.array(X_test), np.array(y_test)

# dar la forma requerida a los datos de entrenamiento(batch size, number of timestamps, number of indicators)
X_train = np.reshape(X_train, newshape = (X_train.shape[0], X_train.shape[1], len(features)))
X_test = np.reshape(X_test, newshape = (X_test.shape[0], X_test.shape[1], len(features)))
```

### Creación y entrenamiento

```python
def build_rnn(input_shape):
    model = Sequential()
    model.add(SimpleRNN(units = 30, return_sequences = True, input_shape=input_shape))
    model.add(Dropout(rate=0.2))
    model.add(SimpleRNN(units = 30, return_sequences = True))
    model.add(Dropout(rate=0.2))
    model.add(SimpleRNN(units = 30, return_sequences = False))
    model.add(Dropout(rate=0.2))
    model.add(Dense(units = 1))

    model.compile(optimizer = 'adam', loss = 'mean_squared_error')
    
    return model


def build_gru(input_shape):
    # crear modelo GRU
    model = Sequential()
    model.add(GRU(units = 50, return_sequences = True, input_shape=input_shape))
    model.add(Dropout(rate=0.2))
    model.add(GRU(units = 50, return_sequences = True))
    model.add(Dropout(rate=0.2))
    model.add(GRU(units = 50, return_sequences = False))
    model.add(Dropout(rate=0.2))
    model.add(Dense(units = 1))

    model.compile(optimizer = 'adam', loss = 'mean_squared_error')

    return model


def build_lstm(input_shape):
    # crear el modelo LSTM
    model = Sequential()
    model.add(LSTM(units = 100, return_sequences = True, input_shape=input_shape))
    model.add(Dropout(rate=0.2))
    model.add(LSTM(units = 100, return_sequences = True))
    model.add(Dropout(rate=0.2))
    model.add(LSTM(units = 100, return_sequences = False))
    model.add(Dropout(rate=0.2))
    model.add(Dense(units = 1))

    model.compile(optimizer = 'adam', loss = 'mean_squared_error')

    return model

models = [
    ('RNN', build_rnn((X_train.shape[1], len(features)))),
    ('GRU', build_gru((X_train.shape[1], len(features)))),
    ('LSTM', build_lstm((X_train.shape[1], len(features))))
]

histories = []

for name, model in models:
    histories.append(model.fit(x = X_train, y = y_train, epochs = 30, verbose=0))
```

```python
# comprobar loss a lo largo del entrenamiento
plt.figure(figsize=(20, 10))
for history in histories:
    plt.plot(history.history['loss'])
plt.title('Evolución de la función pérdida durante el entrenamiento')
plt.xlabel('Época')
plt.ylabel('Función pérdida')
plt.legend(['RNN', 'GRU', 'LSTM'])

save_fig('figure_4_1')
plt.show()
```

### Testeo

```python
# predecir
predictions = []

for name, model in models:
    predictions.append((name,
                        scaler.inverse_transform(np.concatenate((model.predict(X_test), np.zeros((len(X_test), len(features)-1))), axis=1))[:, [0]],
                        model.evaluate(X_test, y_test, verbose = 0)
                       ))
    
y_test = scaler.inverse_transform(np.concatenate((y_test.reshape(-1, 1), np.zeros((len(X_test), len(features)-1))), axis=1))[:, [0]]
```

```python
# evaluar predicciones
for name, y_pred, loss in predictions:
    print('\n%s:' %(name))
    print('\tError cuadrático medio: ', mean_squared_error(y_test, y_pred))
    print('\tCoeficiente de determinación: ', r2_score(y_test, y_pred))
    print('\tFunción pérdida: ', loss)

print('\nDato del día anterior:')
print('\tError cuadrático medio: ', mean_squared_error(y_test[:-1], y_test[1:]))
print('\tCoeficiente de determinación: ', r2_score(y_test[:-1], y_test[1:]))

# visualizar las predicciones
plt.figure(figsize=(20, 10))
plt.plot(y_test, color = 'red', label = 'Precio real')
for name, y_pred, loss in predictions:
    plt.plot(y_pred, label = 'Precio predicho por %s' %(name))
plt.title('Predicción del precio de cierre')
plt.xlabel('Tiempo')
plt.ylabel('Precio')
plt.legend()

save_fig('figure_4_2')
plt.show()
```

```python
features.append(target)

# obtener conjuntos de entrenamiento y testeo como array de numpy bidimensiaonal
# al de testeo se añaden al principio los últimos del entrenamiento porque se necesitan lookback para predecir
training_set = stock_prices.loc[:test_date, features].values
testing_set = np.concatenate((training_set[-lookback+1:],
                                  stock_prices.loc[test_date:, features].values))

# escalar variables
scaler = MinMaxScaler()
training_set = scaler.fit_transform(training_set)
testing_set = scaler.transform(testing_set)

# crear sliding window (ventana de predicción)
X_train = []
y_train = []
for i in range(lookback-1, len(training_set)):
    X_train.append(training_set[i-lookback+1: i+1, :-1])
    # en test se añade solo el target
    y_train.append(training_set[i, -1])
X_train, y_train = np.array(X_train), np.array(y_train)

# crear sliding window de testeo
X_test = []
y_test = []
for i in range(lookback-1, len(testing_set)): 
    X_test.append(testing_set[i-lookback+1: i+1, :-1])
    y_test.append(testing_set[i, -1])
X_test, y_test = np.array(X_test), np.array(y_test)

# dar la forma requerida a los datos de entrenamiento(batch size, number of timestamps, number of indicators)
X_train = np.reshape(X_train, newshape = (X_train.shape[0], X_train.shape[1], len(features) - 1))
X_test = np.reshape(X_test, newshape = (X_test.shape[0], X_test.shape[1], len(features) - 1))
```

### Creación y entrenamiento

```python
def build_rnn(input_shape):
    # crear el modelo RNN
    model = Sequential()
    model.add(SimpleRNN(units = 30, return_sequences = True, input_shape=input_shape))
    model.add(BatchNormalization())
    model.add(SimpleRNN(units = 30, return_sequences = True))
    model.add(BatchNormalization())
    model.add(SimpleRNN(units = 30, return_sequences = False))
    model.add(BatchNormalization())
    model.add(Dense(units = 1))

    model.compile(optimizer = 'adam', loss = 'mean_squared_error')
    
    return model


def build_gru(input_shape):
    # crear modelo GRU
    model = Sequential()
    model.add(GRU(units = 50, return_sequences = True, input_shape=input_shape))
    model.add(BatchNormalization())
    model.add(GRU(units = 50, return_sequences = True))
    model.add(BatchNormalization())
    model.add(GRU(units = 50, return_sequences = False))
    model.add(BatchNormalization())
    model.add(Dense(units = 1))

    model.compile(optimizer = 'adam', loss = 'mean_squared_error')

    return model


def build_lstm(input_shape):
    # crear el modelo LSTM
    model = Sequential()
    model.add(LSTM(units = 100, return_sequences = True, input_shape=input_shape))
    model.add(BatchNormalization())
    model.add(LSTM(units = 100, return_sequences = True))
    model.add(BatchNormalization())
    model.add(LSTM(units = 100, return_sequences = False))
    model.add(BatchNormalization())
    model.add(Dense(units = 1))

    model.compile(optimizer = 'adam', loss = 'mean_squared_error')

    return model

models = [
    ('RNN', build_rnn((X_train.shape[1], len(features) - 1))),
    ('GRU', build_gru((X_train.shape[1], len(features) - 1))),
    ('LSTM', build_lstm((X_train.shape[1], len(features) - 1)))
]

histories = []

for name, model in models:
    histories.append(model.fit(x = X_train, y = y_train, epochs = 30, verbose=0))
```

```python
# comprobar loss a lo largo del entrenamiento
plt.figure(figsize=(20, 10))
for history in histories:
    plt.plot(history.history['loss'])
plt.title('Evolución de la función pérdida durante el entrenamiento')
plt.xlabel('Época')
plt.ylabel('Función pérdida')
plt.legend(['RNN', 'GRU', 'LSTM'])

save_fig('figure_4_3')
plt.show()
```

```python
# predecir
predictions = []

for name, model in models:
    predictions.append((name,
                        scaler.inverse_transform(np.concatenate((model.predict(X_test), np.zeros((len(X_test), len(features)-1))), axis=1))[:, [0]],
                        model.evaluate(X_test, y_test, verbose = 0)
                       ))
    
y_test = scaler.inverse_transform(np.concatenate((y_test.reshape(-1, 1), np.zeros((len(X_test), len(features)-1))), axis=1))[:, [0]]
```

```python
# evaluar predicciones
for name, y_pred, loss in predictions:
    print('\n%s:' %(name))
    print('\tError cuadrático medio: ', mean_squared_error(y_test, y_pred))
    print('\tCoeficiente de determinación: ', r2_score(y_test, y_pred))
    print('\tFunción pérdida: ', loss)

print('\nDato anterior en el horizonte temporal:')
print('\tError cuadrático medio: ', mean_squared_error(y_test[:-5], y_test[5:]))
print('\tCoeficiente de determinación: ', r2_score(y_test[:-5], y_test[5:]))

# visualizar las predicciones
plt.figure(figsize=(20, 10))
plt.plot(y_test, color = 'red', label = 'Precio real')
for name, y_pred, loss in predictions:
    plt.plot(y_pred, label = 'Precio predicho por %s' %(name))
plt.title('Predicción del precio de cierre')
plt.xlabel('Tiempo')
plt.ylabel('Precio')
plt.legend()

save_fig('figure_4_4')
plt.show()
```

# 3. Predecir los retornos
En este apartado se intenta predecir el cambio del precio en un horizonte temporal, es decir, los retornos futuros, representado en la variable 'returns' y 'future_returns'. Esta es una variable cuantitativa, continua y estacionaria.

## 3.1. Retornos en el horizonte temporal

### Definir las variables necesarias para el modelo

```python
# variables independientes que se utilizan para la predicción
features = ['volume', 'adx', 'nvi', 'smad_10', 'smad_20', 'rsi_14', 'rsi_28', 'bbd', 'ppo']

# variable objetivo
target = 'future_returns'

# fecha de testeo
test_date = '2018-1-1'
```

### Preprocesado

```python
# obtener conjuntos de entrenamiento y testeo
X_train = stock_prices.loc[:test_date, features]
X_test = stock_prices.loc[test_date:, features]
y_train = stock_prices.loc[:test_date, target]
y_test = stock_prices.loc[test_date:, target]

# escalar las variables
sc_X = StandardScaler()
X_train = sc_X.fit_transform(X_train)
X_test = sc_X.transform(X_test)

sc_y = StandardScaler()
y_train = sc_y.fit_transform(y_train.values.reshape(-1, 1))
y_test = sc_y.transform(y_test.values.reshape(-1, 1))
```

### Entrenamiento y testeo

```python
models = [
    ('Dummy', DummyRegressor(strategy="mean")),
    ('LR', LinearRegression()),
    ('SVR', SVR(kernel = 'rbf')),
    ('RF', RandomForestRegressor(n_estimators = 200)),
    ('XGB', XGBRegressor(objective ='reg:squarederror', n_estimators=200, learning_rate=0.01))
]
# por cada modelo
for name, model in models:
    # entrenarlo
    model.fit(X_train, y_train.ravel())
    
    # predecir conjunto de testeo
    y_pred = model.predict(X_test)
        
    # evaluar error
    print(name, ':')
    print('\tError cuadrático medio: %.5f' % mean_squared_error(y_test, sc_y.inverse_transform(y_pred.reshape(-1, 1))))
    print('\tCoeficiente de determinación: %.5f'% model.score(X_test, y_test))
            
```

```python
features.append(target)

# obtener conjuntos de entrenamiento y testeo como array de numpy bidimensiaonal
# al de testeo se añaden al principio los últimos del entrenamiento porque se necesitan lookback para predecir
training_set = stock_prices.loc[:test_date, features].values
testing_set = np.concatenate((training_set[-lookback+1:],
                                  stock_prices.loc[test_date:, features].values))

# escalar variables
scaler = MinMaxScaler()
training_set = scaler.fit_transform(training_set)
testing_set = scaler.transform(testing_set)

# crear sliding window (ventana de predicción)
X_train = []
y_train = []
for i in range(lookback-1, len(training_set)):
    X_train.append(training_set[i-lookback+1: i+1, :-1])
    # en test se añade solo el target
    y_train.append(training_set[i, -1])
X_train, y_train = np.array(X_train), np.array(y_train)

# crear sliding window de testeo
X_test = []
y_test = []
for i in range(lookback-1, len(testing_set)): 
    X_test.append(testing_set[i-lookback+1: i+1, :-1])
    y_test.append(testing_set[i, -1])
X_test, y_test = np.array(X_test), np.array(y_test)

# dar la forma requerida a los datos de entrenamiento(batch size, number of timestamps, number of indicators)
X_train = np.reshape(X_train, newshape = (X_train.shape[0], X_train.shape[1], len(features) - 1))
X_test = np.reshape(X_test, newshape = (X_test.shape[0], X_test.shape[1], len(features) - 1))
```

```python
def build_rnn(input_shape):
    # crear el modelo RNN
    model = Sequential()
    model.add(SimpleRNN(units = 30, return_sequences = True, input_shape=input_shape))
    model.add(Dropout(rate=0.2))
    model.add(SimpleRNN(units = 30, return_sequences = True))
    model.add(Dropout(rate=0.2))
    model.add(SimpleRNN(units = 30, return_sequences = False))
    model.add(Dropout(rate=0.2))
    model.add(Dense(units = 1))

    model.compile(optimizer = 'adam', loss = 'mean_squared_error')
    
    return model


def build_gru(input_shape):
    # crear modelo GRU
    model = Sequential()
    model.add(GRU(units = 50, return_sequences = True, input_shape=input_shape))
    model.add(Dropout(rate=0.2))
    model.add(GRU(units = 50, return_sequences = True))
    model.add(Dropout(rate=0.2))
    model.add(GRU(units = 50, return_sequences = False))
    model.add(Dropout(rate=0.2))
    model.add(Dense(units = 1))

    model.compile(optimizer = 'adam', loss = 'mean_squared_error')

    return model


def build_lstm(input_shape):
    # crear el modelo LSTM
    model = Sequential()
    model.add(LSTM(units = 100, return_sequences = True, input_shape=input_shape))
    model.add(Dropout(rate=0.2))
    model.add(LSTM(units = 100, return_sequences = True))
    model.add(Dropout(rate=0.2))
    model.add(LSTM(units = 100, return_sequences = False))
    model.add(Dropout(rate=0.2))
    model.add(Dense(units = 1))

    model.compile(optimizer = 'adam', loss = 'mean_squared_error')

    return model

models = [
    ('RNN', build_rnn((X_train.shape[1], len(features) - 1))),
    ('GRU', build_gru((X_train.shape[1], len(features) - 1))),
    ('LSTM', build_lstm((X_train.shape[1], len(features) - 1)))
]

histories = []

for name, model in models:
    histories.append(model.fit(x = X_train, y = y_train, epochs = 30, verbose=0))
```

```python
# obtener conjuntos de entrenamiento y testeo como array de numpy bidimensiaonal
# al de testeo se añaden al principio los últimos del entrenamiento porque se necesitan lookback para predecir
training_set = stock_prices.loc[:test_date, features].values
testing_set = np.concatenate((training_set[-lookback:],
                                  stock_prices.loc[test_date:, features].values))
target_training_set = stock_prices.loc[:test_date, target].values
target_testing_set = np.concatenate((target_training_set[-lookback:],
                                  stock_prices.loc[test_date:, target].values))

# escalar variables
scaler = MinMaxScaler()
training_set = scaler.fit_transform(training_set)
testing_set = scaler.transform(testing_set)

# transformar target en vector
enc = OneHotEncoder(handle_unknown='ignore')
target_training_set = enc.fit_transform(target_training_set.reshape(-1, 1)).toarray()
target_testing_set = enc.transform(target_testing_set.reshape(-1, 1)).toarray()

# crear sliding window (ventana de predicción)
X_train = []
y_train = []
for i in range(lookback, len(training_set) + 1):
    X_train.append(training_set[i-lookback: i, :])
    # en test se añade solo el target
    y_train.append(target_training_set[i-1])
X_train, y_train = np.array(X_train), np.array(y_train)

# crear sliding window de testeo
X_test = []
y_test = []
for i in range(lookback, len(testing_set) + 1): 
    X_test.append(testing_set[i-lookback: i, :])
    y_test.append(target_testing_set[i-1])
X_test, y_test = np.array(X_test), np.array(y_test)

# dar la forma requerida a los datos de entrenamiento(batch size, number of timestamps, number of indicators)
X_train = np.reshape(X_train, newshape = (X_train.shape[0], X_train.shape[1], len(features)))
X_test = np.reshape(X_test, newshape = (X_test.shape[0], X_test.shape[1], len(features)))
```

### Creación y entrenamiento

```python
def build_rnn(input_shape):
    # crear el modelo RNN
    model = Sequential()
    model.add(SimpleRNN(units = 30, return_sequences = True, input_shape=input_shape))
    model.add(Dropout(rate=0.2))
    model.add(SimpleRNN(units = 30, return_sequences = True))
    model.add(Dropout(rate=0.2))
    model.add(SimpleRNN(units = 30, return_sequences = False))
    model.add(Dropout(rate=0.2))
    model.add(Dense(units = 3, activation='softmax'))

    model.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics=['accuracy'])
    
    return model


def build_gru(input_shape):
    # crear modelo GRU
    model = Sequential()
    model.add(GRU(units = 50, return_sequences = True, input_shape=input_shape))
    model.add(Dropout(rate=0.2))
    model.add(GRU(units = 50, return_sequences = True))
    model.add(Dropout(rate=0.2))
    model.add(GRU(units = 50, return_sequences = False))
    model.add(Dropout(rate=0.2))
    model.add(Dense(units = 3, activation='softmax'))

    model.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics=['accuracy'])

    return model


def build_lstm(input_shape):
    # crear el modelo LSTM
    model = Sequential()
    model.add(LSTM(units = 100, return_sequences = True, input_shape=input_shape))
    model.add(Dropout(rate=0.2))
    model.add(LSTM(units = 100, return_sequences = True))
    model.add(Dropout(rate=0.2))
    model.add(LSTM(units = 100, return_sequences = False))
    model.add(Dropout(rate=0.2))
    model.add(Dense(units = 3, activation='softmax'))

    model.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics=['accuracy'])

    return model

models = [
    ('RNN', build_rnn((X_train.shape[1], len(features)))),
    ('GRU', build_gru((X_train.shape[1], len(features)))),
    ('LSTM', build_lstm((X_train.shape[1], len(features))))
]

histories = []

for name, model in models:
    histories.append(model.fit(x = X_train, y = y_train, epochs = 30, verbose=0))
```

### Testeo

```python
for name, model in models:
    print('\n%s:' %(name))

    # predecir target y medir precision
    y_pred = np.argmax(model.predict(X_train), axis=1)
    print('\tPrecisión en el conjunto de entrenamiento: {0:.2%}'.format(np.equal(y_pred, np.argmax(y_train, axis=1)).mean()))
    
    y_pred = np.argmax(model.predict(X_test), axis=1)
    print('\tPrecisión en el conjunto de testeo: {0:.2%}'.format(np.equal(y_pred, np.argmax(y_test, axis=1)).mean()))

    # matriz de confusión
    print(confusion_matrix(np.argmax(y_test, axis=1), y_pred))
```
